```python
output = ""
output += "ClrList L1\n"
for conv_id in range(conv_weights.shape[-1]):
    output += "For(J,1,28,4\n"
    output += "For(I,1,28,4\n"
    # output += "0->A"
    for i in range(4):
        for j in range(4):
            # wsum += img[0, y+i, x+j, 0] * conv_weights[i, j, 0]
            output += "max(0,"
            weight = round(float(conv_weights[i, j, conv_id]), 3)
            if weight != 0:
                # print(weight)
                output += f"pxl-Test(I+{i},J+{j})*{tinbasic.needcast(weight)}+"
    weight = round(float(conv_bias[conv_id]), 3)
    if output[-1] == "+":
        if weight != 0 and weight > 0:
                output += str(weight)
        elif weight !=0 and weight < 0:
            output = output[:-1] + f"-{str(abs(weight))}"
        else:
            output = output[:-1]
    else:
        output += tinbasic.needcast(weight)

    output += "->L1(1+dim(L1))\n"
    output += "End\n"*2
# print(output)
```
